```python
# Peforming exploratory data analysis on the netflix_data.csv data to understand what may be contributing to movies getting shorter over time.

# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Load and inspect the Netflix data
netflix_df = pd.read_csv('netflix_data.csv', index_col=0)

# Subset the DataFrame for "Movies"
netflix_subset = netflix_df[netflix_df["type"] == "Movie"]

# Subset the netflix_subset to only the columns title, country, genre, release_year, and duration
netflix_movies = netflix_subset.loc[:, ["title", "country", "genre", "release_year", "duration"]]

# Filter the DataFrame by Movie duration
short_movies = netflix_movies[netflix_movies["duration"]<60]

# Assign colors to movie genres
colors = []
for index, row in netflix_movies.iterrows():
    if row['genre'] == "Children":
        colors.append("yellow")
    elif row['genre'] == "Documentaries":
        colors.append("green")
    elif row['genre'] == "Stand-Up":
        colors.append("blue")
    else:
        colors.append("red")

# Plot the movie duration by year using the genre colors created
fig = plt.figure(figsize=(12, 8))
plt.scatter(netflix_movies['release_year'], netflix_movies['duration'], c=colors)
plt.xlabel("Release year") 
plt.ylabel("Duration (min)")
plt.title("Movie Duration by Year of Release")
plt.show()

# Answer "Are we certain that movies are getting shorter?"
answer = "no"
```
