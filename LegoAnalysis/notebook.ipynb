```python
# Import pandas
import pandas as pd

# Read colors data
colors = pd.read_csv('datasets/colors.csv')

# Print the first few rows
print(colors.head(5))
```

```python
# How many distinct colors are available?
num_colors = colors['name'].nunique()
print(num_colors)
```

```python
# colors_summary: Distribution of colors based on transparency
colors_summary = colors.groupby('is_trans').count()
print(colors_summary)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Read sets data as `sets`
sets = pd.read_csv('datasets/sets.csv')

# Create a summary of average number of parts by year: `parts_by_year`
# parts_by_year = sets[['year','num_parts']].groupby('year').mean().round(2)
parts_by_year = sets.groupby('year')['num_parts'].mean().round(2)
print(parts_by_year)
# Plot trends in average number of parts by year
parts_by_year.plot()
plt.show()
```

```python
# themes_by_year: Number of themes shipped by year
themes_by_year = sets.groupby('year')['theme_id'].nunique()
print(themes_by_year)
print(type(themes_by_year))
```

```python
# Get the number of unique themes released in 1999
# num_themes = themes_by_year[themes_by_year['year']=='1999']
num_themes = themes_by_year.loc[1999]
print(num_themes)
```
